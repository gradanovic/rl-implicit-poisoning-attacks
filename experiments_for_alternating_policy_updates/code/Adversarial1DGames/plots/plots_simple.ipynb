```python
import matplotlib.pyplot as plt
import json

plt.rcParams["axes.facecolor"] = "lightgrey"
```

```python
with open("results.json", "r") as f:
    dat = json.load(f)
print(dat.keys())
plt.figure(figsize= (20,10))
plt.plot(dat["alphas"], dat["costs"], "--bo", label = "cost: 1 - $\\pi_{{learner}}(a^{{\\dagger}})$")
plt.plot(dat["alphas"], dat["distances"], "--ro", label = "distance : $kl_div(\\pi_1, \\pi_1^0)$")
plt.plot(dat["alphas"], [c + a * d for c,a,d in zip(dat["costs"], dat["alphas"], dat["distances"])], "--ko", label = "cost + $\\alpha * distance$")
plt.legend()
plt.grid()
plt.xlabel("$\\alpha$")
plt.ylabel("cost/distance")
plt.title("Cost and distance on the simple 1D game using alternating optimization with PPO")
```
